```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a random array of lowercase alphabets
random_alphabets = np.random.choice(list('abcdefghijklmnopqrstuvwxyz'), size=1000)

# Count the occurrences of each alphabet
alphabet_counts = {char: np.sum(random_alphabets == char) for char in set(random_alphabets)}

# Sort the alphabet counts by their occurrences
sorted_alphabet_counts = sorted(alphabet_counts.items(), key=lambda x: x[1])

# Extract the sorted characters and counts
sorted_characters = [item[0] for item in sorted_alphabet_counts]
sorted_counts = [item[1] for item in sorted_alphabet_counts]

# Plot the distribution in a bar chart
plt.figure(figsize=(10, 6))
plt.bar(range(len(sorted_characters)), sorted_counts, align='center', color='skyblue')
plt.xticks(range(len(sorted_characters)), sorted_characters)
plt.xlabel('Alphabets')
plt.ylabel('Count')
plt.title('Distribution of Alphabets')
plt.show()
```
